### 1. Loss Function (or Error Function)

A **Loss Function** measures the error for a *single training example*. It quantifies how poorly a single model prediction performs compared to the actual target value. The goal during training is to minimize this loss.

Here's an example using Mean Squared Error (MSE), a common loss function for regression tasks. It calculates the average of the squares of the errors between actual and predicted values.

```python
import numpy as np

def mean_squared_error_loss(y_true, y_pred):
    """
    Calculates the Mean Squared Error (MSE) loss for a single example.
    y_true: actual target value
    y_pred: predicted value
    """
    return (y_true - y_pred)**2

# Example usage for a single prediction
y_actual = 10
y_predicted = 8
loss = mean_squared_error_loss(y_actual, y_predicted)
print(f"Actual: {y_actual}, Predicted: {y_predicted}, Loss (MSE): {loss}")

y_actual = 10
y_predicted = 10.5
loss = mean_squared_error_loss(y_actual, y_predicted)
print(f"Actual: {y_actual}, Predicted: {y_predicted}, Loss (MSE): {loss}")
```

### 2. Cost Function (or Objective Function)

A **Cost Function** (also known as an Objective Function) measures the overall error or performance of the model across the *entire training dataset* (or a mini-batch). It is typically the average of the loss functions over all training examples. The ultimate goal of training a neural network is to minimize this cost function through techniques like gradient descent.

Using the Mean Squared Error as a loss function, the corresponding cost function would be the average MSE over all examples.

```python
import numpy as np

def mean_squared_error_cost(y_true_batch, y_pred_batch):
    """
    Calculates the Mean Squared Error (MSE) cost for a batch of examples.
    y_true_batch: array of actual target values
    y_pred_batch: array of predicted values
    """
    losses = (y_true_batch - y_pred_batch)**2
    return np.mean(losses)

# Example usage for a batch of predictions
y_actual_batch = np.array([10, 20, 30, 40])
y_predicted_batch = np.array([8, 22, 28, 41])
cost = mean_squared_error_cost(y_actual_batch, y_predicted_batch)
print(f"Actual batch: {y_actual_batch}")
print(f"Predicted batch: {y_predicted_batch}")
print(f"Cost (Average MSE for batch): {cost}")
```

### 3. Activation Function

**Activation Functions** introduce non-linearity into the neural network, allowing it to learn complex patterns and relationships in the data. Without activation functions, a neural network would simply be a linear regression model, regardless of how many layers it has.

Each neuron in a hidden layer applies an activation function to its weighted sum of inputs and bias. Here are some common activation functions:

*   **ReLU (Rectified Linear Unit)**: `f(x) = max(0, x)`. Popular due to its simplicity and effectiveness in combating vanishing gradients.
*   **Sigmoid**: `f(x) = 1 / (1 + exp(-x))`. Squashes values between 0 and 1, often used in the output layer for binary classification.
*   **Tanh (Hyperbolic Tangent)**: `f(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))`. Squashes values between -1 and 1, similar to sigmoid but centered around 0.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

# Generate a range of input values
x = np.linspace(-5, 5, 100)

# Calculate outputs for each activation function
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Plotting the activation functions
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x, y_relu)
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x, y_sigmoid)
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x, y_tanh)
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('Tanh(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
